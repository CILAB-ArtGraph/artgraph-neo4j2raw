```python
from ArtGraphOGM import  ArtGraphDBConnector
from dotenv import load_dotenv, find_dotenv
import os
```

```python
#get db credentials from the .env file
dotenv_path = find_dotenv()
load_dotenv(dotenv_path)
```

```python
conf = {
    'uri': os.environ.get("uri"),
    'username': os.environ.get("username"),
    'password': os.environ.get("password"),
    'database': os.environ.get("database")
}
```

```python
agdb = ArtGraphDBConnector(conf['uri'], conf['username'], conf['password'], conf['database'])
```

```python
#node count
node_count_query = "MATCH (n) RETURN distinct labels(n) as node_label, count(*) as count"
df_node_count = agdb.df_from_query(node_count_query)
```

```python
df_node_count
```

```python
#relations
relation_query = "MATCH (n)-[r]->(n2) RETURN distinct type(r) as rel_label, count(*) as count"
df_relations = agdb.df_from_query(relation_query)
```

```python
df_relations
```

```python
#triple list
triple_list_query = "MATCH (x)-[r]->(y) RETURN distinct HEAD(labels(x)) as head, type(r), head(labels(y)) as tail"
df_triple_list = agdb.df_from_query(triple_list_query)
```

```python
df_triple_list
```

```python
#count artworks by genre
artworks_by_genre_query = "MATCH (n:Artwork)-[:Genre]->(g:Genre) RETURN distinct(g.name) as genre, count(*) as count ORDER BY count DESC"
artworks_by_genre = agdb.df_from_query(artworks_by_genre_query)
```

```python
artworks_by_genre
```

```python
#count artworks by style
artworks_by_style_query = "MATCH (n:Artwork)-[:Style]->(s:Style) RETURN distinct(s.name) as style, count(*) as count ORDER BY count DESC"
artworks_by_style = agdb.df_from_query(artworks_by_style_query)
```

```python
artworks_by_style
```

```python
#count artworks by author
artworks_by_artist_query = "MATCH (n:Artwork)-[:Author]->(a:Artist) RETURN distinct(a.name) as artist, count(*) as count ORDER BY count DESC"
artworks_by_artist = agdb.df_from_query(artworks_by_artist_query)
```
